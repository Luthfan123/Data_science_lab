```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
x=np.arange(0,90,10)
y=np.sin(x)
plt.plot(x,y,'g')
plt.xlabel("x axix")
plt.ylabel("y axis")
plt.title("sin(x) graph")
```

```python
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
x=np.arange(0,4*np.pi)
y=np.sin(x)
z=np.cos(x)
plt.plot(x,y,'g*')
plt.plot(x,z,'b-')
plt.xlabel("x axix")
plt.ylabel("y axis")
plt.title("sin(x)-cos(x) graph")
plt.legend(['sin','cos'])
```
