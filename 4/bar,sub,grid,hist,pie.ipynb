```python
import numpy as np
from matplotlib import pyplot as plt
x=np.array([2,4,6,8])
y=np.array([2,4,6,8])
plt.bar(x,y,width=1.2)
plt.show()
```

```python
plt.barh(x,y,height=1.2)
plt.show()
```

```python
x=[1,4,7,20]
y=[2,5,8,11]
plt.subplot(5,5,1)
plt.plot(x,y)
plt.show()
x=[1,4,7,20]
y=[11,8,5,2]
plt.subplot(5,5,2)
plt.plot(x,y)
plt.show()
```

```python
plt.grid()
x=[1,4,7,10]
y=[2,5,8,11]
plt.plot(x,y)
plt.show()
```

```python

x=np.random.normal(100,2,250)
plt.hist(x)
plt.show()
```

```python
from matplotlib import pyplot as plt
x=[2,4,6,8]
y=[2,5,8,11]
plt.plot(x,y,linewidth=1.8,linestyle='dashed')
plt.show()
```

```python
from matplotlib import pyplot as plt
la=['apple','orange','banana']
x=[2,4,6]
plt.pie(x,labels=la)
plt.show()
```

```python
from matplotlib import pyplot as plt
la=['apple','orange','banana']
x=[2,4,6]
ex=[0.2,0,0]
plt.pie(x,labels=la,explode=ex,shadow=True)
plt.show()
```
