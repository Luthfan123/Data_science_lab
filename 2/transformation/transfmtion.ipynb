```python
#rotate
from matplotlib import pyplot as plt
import numpy as np
from PIL import Image
from scipy import ndimage
a=Image.open("tree.jpeg")
b=np.array(a)
c=ndimage.rotate(b,45,cval=128)
d=Image.fromarray(c)
plt.imshow(d)

```

```python
#scale
from matplotlib import pyplot as plt
import numpy as np
from PIL import Image
from scipy import ndimage
a=Image.open("tree.jpeg")
b=np.array(a)
c=ndimage.zoom(b,(0.5,0.5,1))
d=Image.fromarray(c)
plt.imshow(d)
```
